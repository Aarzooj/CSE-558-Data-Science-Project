```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
original = [91.409, 91.667, 88.354]
scaling = [91.18, 91.31, 86.717]

models = ['Random Forest', 'XGBoost', 'Logistic Regression']
bar_width = 0.35

index = np.arange(len(models))

plt.figure(figsize=(10, 5))
bars1 = plt.bar(index - bar_width / 2, original, bar_width, label='Original Dataset', color='skyblue')
bars2 = plt.bar(index + bar_width / 2, scaling, bar_width, label='After Feature Hashing', color='orange')

# Add annotations
for bar in bars1:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():.2f}',
        ha='center',
        va='bottom'
    )

for bar in bars2:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():.2f}',
        ha='center',
        va='bottom'
    )

plt.xlabel('Models')
plt.ylabel('Accuracies')
plt.title('Accuracy of Models on Original and Feature Hashed Datasets')
plt.xticks(index, models)
plt.legend()
plt.grid(False)

plt.show()
```
